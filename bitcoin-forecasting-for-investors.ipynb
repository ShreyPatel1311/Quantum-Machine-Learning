```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from prophet import Prophet
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.layers import Layer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, GRU
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, LearningRateScheduler, TerminateOnNaN
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.optimizers.schedules import ExponentialDecay
from tensorflow.keras.models import Model

import warnings
warnings.filterwarnings('ignore')
```

```python
def preprocessing(dataset, date_str, y_str, vol_str):
    dataset[date_str] = pd.to_datetime(dataset[date_str], errors='coerce', utc=True)
    dataset = dataset.set_index(date_str)
    dataset = dataset.resample('1H').agg({'Close':'mean', 'Volume':'mean'})
    dataset = dataset.reset_index()
    dataset = dataset[[date_str, y_str, vol_str]].copy()
    dataset = dataset.rename(columns={date_str: 'ds', y_str: 'y', vol_str: 'volume'})
    dataset['ds'] = pd.to_datetime(dataset['ds'], errors='coerce').dt.tz_localize(None)
    dataset['y'] = dataset['y'].astype(float)
    dataset.dropna(inplace=True)

    return dataset
```

```python
stock_df = pd.read_csv('E:\\College\\College\\Quantum Computing\\World-Stock-Prices-Dataset.csv')
stock_df = stock_df[stock_df['Ticker']=='NVDA']
sentiment_df = pd.read_csv("out.csv")
stock_df = preprocessing(stock_df, 'Date', 'Close', 'Volume')
```

```python
sentiment_df['ds'] = pd.to_datetime(sentiment_df['ds'], utc=True)
stock_df['ds'] = pd.to_datetime(stock_df['ds'], utc=True)
stock_df['ds'].dtype, sentiment_df['ds'].dtype
```

```python
sentiment_avg = sentiment_df.groupby('ds', as_index=False)['score'].mean()
stock_df = pd.merge(stock_df, sentiment_avg, on='ds', how='left')
stock_df = stock_df.fillna(0)
```

```python
stock_df['ds'] = pd.to_datetime(stock_df['ds']).dt.tz_localize(None)
```

```python
model = Prophet(seasonality_mode="additive", seasonality_prior_scale=0.01, changepoint_prior_scale=0.1, changepoint_range=0.98, interval_width=0.9, uncertainty_samples=4500, n_changepoints=30)
model.add_seasonality(name='monthly', period=30.5, fourier_order=5, prior_scale=35)
model.add_seasonality(name='yearly', period=7, fourier_order=5)
model.fit(stock_df[6000:])
```

```python
future = model.make_future_dataframe(periods=90)
```

```python
forecast = model.predict(future)
```

```python
def create_features(df):
    df = df.copy()
    df['ds'] = pd.to_datetime(df['ds'], errors='coerce', utc=True).dt.tz_localize(None)
    df["SMA_100"] = df["y"].rolling(window=100).mean()
    df["EMA_100"] = df["y"].ewm(span=100, adjust=False).mean()
    df["rolling_std_100"] = df["y"].rolling(window=100).std()

    df["bollinger_upper"] = df["SMA_100"] + (df["rolling_std_100"] * 2)
    df["bollinger_lower"] = df["SMA_100"] - (df["rolling_std_100"] * 2)
    df['volume'] = np.log(df['volume'])
    
    df['yhat'] = df['yhat']
    return df
```

```python
xgb_train = create_features(stock_df)
xgb_train['yhat_upper'] = forecast['yhat_upper']
xgb_train['yhat_lower'] = forecast['yhat_lower']
X_train = xgb_train[['SMA_100', 'EMA_100', 'rolling_std_100', 'bollinger_upper', 'bollinger_lower', 'score', 'yhat', 'yhat_upper', 'yhat_lower']]
y_train = xgb_train['residual']
```

```python
xgb_train.isna().sum()
```

```python
model_xgb = xgb.XGBRegressor(n_estimators=400, max_depth=100, learning_rate=0.01)
model_xgb.fit(X_train, y_train)
```

```python
future = model.make_future_dataframe(periods=365*24, freq='H')
forecast = model.predict(future)
forecast = forecast[-365*24:]
xgb_test = create_features(forecast)
X_past_test = xgb_test[['dayofweek', 'day', 'month', 'year', 'weekofyear', 'hour', 'yhat', 'yhat_upper', 'yhat_lower']]
residual_preds = model_xgb.predict(X_past_test)
forecast['yhat_ensemble'] = forecast['yhat'] + residual_preds
```

```python
forecast_test = model.predict(stock_df.iloc[-365*24*4:][['ds']])
forecast_test = forecast_test[-365*24*4:]
xgb_test = create_features(forecast_test)
X_test = xgb_test[['dayofweek', 'day', 'month', 'year', 'weekofyear', 'hour', 'yhat', 'yhat_upper', 'yhat_lower']]
residual_preds = model_xgb.predict(X_test)
forecast_test['yhat_ensemble'] = forecast_test['yhat'] + residual_preds
```
