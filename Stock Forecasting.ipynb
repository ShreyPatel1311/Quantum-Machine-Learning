```python
def create_features(df):
    y_str = "y" if 'y' in df else 'yhat'
    df = df[4000:6729].copy()
    df['ds'] = pd.to_datetime(df['ds'], errors='coerce', utc=True).dt.tz_localize(None)
    df["SMA_100"] = df[y_str].rolling(window=100).mean()
    df["EMA_100"] = df[y_str].ewm(span=100, adjust=False).mean()
    df["rolling_std_100"] = df[y_str].rolling(window=100).std()

    df["bollinger_upper"] = df["SMA_100"] + (df["rolling_std_100"] * 2)
    df["bollinger_lower"] = df["SMA_100"] - (df["rolling_std_100"] * 2)
    return df
```

```python
numQubits = 8
dev = qml.device("lightning.gpu", wires=numQubits)

@qml.qnode(device=dev)
def qkernel_circuit(a, b):
    qml.AngleEmbedding(features=a, wires=range(numQubits), rotation="X")
    qml.adjoint(qml.AngleEmbedding(features=b, wires=range(numQubits), rotation="X"))
    return qml.probs(wires=range(numQubits))
```

```python
def qkernel(A, B):
    return np.array([[qkernel_circuit(a, b)[0]for b in B]for a in A])
```
