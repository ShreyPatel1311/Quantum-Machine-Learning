```python
stocks = pd.read_csv("E:\\College\\College\\Quantum Computing\\Quantum-Machine-Learning\\Stocks\\NVDA.csv")
select_stock = stocks[stocks['Ticker'] == selectedStockTag]
select_stock.head()
```

```python
myStock = select_stock[['Date', 'Close', 'Volume']].copy()
myStock = myStock.rename(columns={'Date': 'ds', 'Close': 'y', 'Volume': 'volume'})
```

```python
scores=[]
dates=[]
for date in myStock['ds']:
    try:
        date_str = pd.to_datetime(date).strftime('%Y-%m-%d')
        res = finnhub_client.company_news(selectedStockTag, _from=date_str, to=date_str)
        sentiment_df = pd.DataFrame(res)
        if not sentiment_df.empty:
            sentiment_df['datetime'] = pd.to_datetime(sentiment_df['datetime'], unit='s')
            sia = SentimentIntensityAnalyzer()
            headlines = [x['headline'] for x in res]
            scores.append(pd.Series([sia.polarity_scores(h)['compound'] for h in headlines]).mean())
            print("went here")
            print(f"{date} : {pd.Series([sia.polarity_scores(h)['compound'] for h in headlines]).mean()}")
        else:
            scores.append(0)
        dates.append(date)
    except:
        print(sentiment_df)
        print(date)
        pass
    time.sleep(0.1)
```
