```python
import pennylane as qml
from pennylane import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
dev = qml.device("default.qubit", wires=64)

def ConvCircuit(theta, wire):
    qml.RZ(-np.pi/2, wires=1)
    qml.CNOT([1, 0])
    qml.RZ(theta[0], wires=0)
    qml.RY(theta[1], wires=1)
    qml.CNOT([0, 1])
    qml.RY(theta[2], 1)
    qml.CNOT([1, 0])
    qml.RZ(-np.pi/2, 0)

@qml.qnode(device=dev)
def testCircuit(params):
    ConvCircuit(params, wire=[0, 1])
    return qml.probs(wires=[0, 1])

qml.draw_mpl(testCircuit, style='black_white')([0.1, 0.2, 0.3])
```
